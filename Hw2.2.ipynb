```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
table = pd.read_csv("2017_jun_final - 2017_jun_final.csv")
```

```python
print(table.head())
```

```python
print(table.shape)
```

```python
print(table.dtypes)
```

```python
sum = np.sum(table.isnull())
print(sum)
```

```python
for row in table:
    if row != "Мова програмування" and table[row].isnull().sum() != 0:
        table = table.drop(columns=[row])
    
```

```python
sum = np.sum(table.isnull())
print(sum)
```

```python
table = table.dropna()
print(table)
```

```python
print(table.shape)
```

```python
print(table["Мова програмування"])
```

```python
python_data = (table[table["Мова програмування"] == "Python"])
print(python_data)
```

```python
print(python_data.shape)
```

```python
grouped = python_data.groupby("Посада")
```

```python
aggregated_df = python_data.groupby("Посада").agg({
    'Зарплата на місяць': ['min', 'max']
}).reset_index()
print(aggregated_df)
```

```python
def fill_avg_salary(group):
    avg_salary = group['Зарплата на місяць'].mean()
    return avg_salary
avg = grouped.apply(fill_avg_salary)
print(avg)
```

```python
summary = avg.describe()
print(summary)
```

```python
summary.to_csv("summary",index = True)
```

```python
print(avg)
print(type(avg))
plt.figure(figsize=(8, 6))
avg.plot(kind = "bar")
plt.xlabel("Посада")
plt.ylabel("Зарплата")
plt.title("Рівні зарплати за посадами")
plt.tight_layout()
```

```python
sr = table.groupby("Мова програмування").size()
plt.figure(figsize=(8, 8))
sr = sr.reset_index(name='кількість')
sr.columns = ["мова програмування","кількість"]
ax = sr.set_index('мова програмування')['кількість'].plot(kind='pie', autopct=None, labels=None,startangle=90)
ax.legend(
    labels=sr['мова програмування'], 
    loc='center left', 
    bbox_to_anchor=(1, 0.5)
)
plt.title('Розподіл мов програмування',size = "25")
plt.ylabel('')
```
